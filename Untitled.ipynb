```python
df = pd.read_csv("Northwind eda dataset c.csv", header=None)
df.head()
```

```python
df.columns = [
    "OrderID",
    "OrderDate",
    "CustomerID",
    "CompanyName",
    "Country",
    "City",
    "Quantity",
    "ProductName",
    "Category",
    "UnitPrice",
    "UnitsInStock",
    "Discount",
    "Revenue"
]

df.head()
```

```python
df.describe()
```

```python
df["Revenue"].sum()
```

```python
df["Category"].value_counts()
```

```python
(
df.groupby("Country")
["Revenue"].sum().sort_values(ascending=False)
)
```

```python
(
df.groupby("Category")
["Revenue"].sum().sort_values(ascending=False)
)
```

```python
top_countries = df.groupby("Country")["Revenue"].sum().sort_values(ascending=False).head(10)
top_countries.plot(kind="bar", figsize=(10,5), title="Top 10 Countries by Revenue")
```

```python
category_revenue = df.groupby("Category")["Revenue"].sum().sort_values(ascending=False)
category_revenue.plot(kind="bar", figsize=(10,5), title="Revenue by Category")
```

```python
df["Category"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(7,7)
)
```

```python
df["Revenue"].plot(
    kind="hist",
    bins=20,
    figsize=(8,5),
    title="Revenue Distribution"
)
```

```python
df.plot(
    x="Quantity",
    y="Revenue",
    kind="scatter",
    figsize=(8,5),
    title="Quantity vs Revenue"
)
```

The dataset contains 2,155 records and 13 columns.
There are no missing values, indicating that the dataset is clean and ready for analysis.
The Beverages category generated the highest total revenue, while Grains/Cereals generated the lowest.
Revenue varies across countries, with some countries contributing significantly more sales than others.
Most transactions generated moderate revenue, while a few generated exceptionally high revenue.
There is a positive relationship between Quantity sold and Revenue, although other factors like Unit Price also affect revenue.
Conclusion
The exploratory data analysis provided valuable insights into sales performance across different countries and product categories. The analysis revealed that Beverages contributed the highest revenue, the dataset contained no missing values, and sales performance differed across regions. These insights can help support business decisions regarding product focus and market strategy.
